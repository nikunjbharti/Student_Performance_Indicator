## We can say that our data is following linear path so we are gonna apply linear regression

```python
df.head()
```

```python
y = df['math_score']
x = df.drop(columns = ['math_score',	'total_score',	'average_score'], axis = 1)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
numeric_feature = x.select_dtypes(exclude = 'object').columns
categorical_feature = x.select_dtypes(include = 'object').columns
preprocessor = ColumnTransformer(
    [
        ('OneHotEncoder', OneHotEncoder(), categorical_feature),
        ('StandardScaler', StandardScaler(), numeric_feature)
    ]
)
```

```python
x_encoded = preprocessor.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_encoded, y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## importing every model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
import warnings
from math import sqrt
# Suppress warnings
warnings.filterwarnings("ignore")
```

```python
model = {
    'Linear Regression' : LinearRegression(),
    'Lasso' : Lasso(),
    'Ridge' : Ridge(),
    'Decision Tree' : DecisionTreeRegressor(),
    'Random Forest' : RandomForestRegressor(),
    'CatBoost' : CatBoostRegressor(verbose=0),
    'XGBoost': XGBRegressor()
}

for key, value in model.items():
  print('-'*40)
  print(f"Model Traning for {key} is going on.........\n")
  value.fit(x_train, y_train)
  print(f"Model Traning for {key} is sucessful\n")
  print('-'*40)
  print('\n')

  y_pred_train = value.predict(x_train)
  y_pred_test = value.predict(x_test)

  ## Model Performanance on traning data

  r_score = r2_score(y_train, y_pred_train)
  mae = mean_absolute_error(y_train, y_pred_train)
  mse = mean_squared_error(y_train, y_pred_train)
  rmse = sqrt(mse)

  print('-'*40)
  print('Model Performance on Train Data')
  print('-'*40)

  print(f"Accurecy of {key} model on training data is : {r_score}")
  print(f"Mean Absolute Error of {key} model on training data is : {mae}")
  print(f"Mean Squared Error of {key} model on training data is : {mse}")
  print(f"Root Mean Squared Error of {key} model on training data is : {rmse}\n")

  print('-'*40)
  print('Model Performance on Test Data')
  print('-'*40)

  ## Model performanance on test data

  r_score_test = r2_score(y_test, y_pred_test)
  mae_test = mean_absolute_error(y_test, y_pred_test)
  mse_test = mean_squared_error(y_test, y_pred_test)
  rmse_test = sqrt(mse_test)


  print(f"Accurecy of {key} model on training data is : {r_score_test}")
  print(f"Mean Absolute Error of {key} model on training data is : {mae_test}")
  print(f"Mean Squared Error of {key} model on training data is : {mse_test}")
  print(f"Root Mean Squared Error of {key} model on training data is : {rmse_test}")

  print('='*40)
  print('\n')
```
